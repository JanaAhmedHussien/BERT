```python
import re
import pyterrier as pt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))
stemmer=PorterStemmer()
```

```python
def Stem_text(text):
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    # print (tokens)
    return ' '.join(stemmed_tokens)

def clean(text):
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=\@]", " ", text) # remove special characters
   text = re.sub(r'\t', ' ', text) # remove tabs
   text = re.sub(r'\n', ' ', text) # remove line jump
   text = re.sub(r"\s+", " ", text) # remove extra white space
   text = text.strip()
   return text

def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words] #Lower is used to normalize al the words make them in lower case
    # print('Tokens are:',tokens,'\n')
    return ' '.join(filtered_tokens)

#we need to process the query also as we did for documents
def preprocess(sentence):
  sentence = clean(sentence)
  sentence = remove_stopwords(sentence)
  sentence = Stem_text(sentence)
  return sentence


cvs['preprocessed_text']=cvs['text'].apply(preprocess)
```

```python
job_description=preprocess(job_description)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def encode(text, max_length=128):
    return bert_tokenizer.encode_plus(
                  text,                      # Sentence to encode.
                  add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                  truncation=True,
                  max_length = max_length,           # Pad & truncate all sentences.
                  padding="max_length",
                  return_attention_mask = True,   # Construct attention mask
                  return_tensors = 'pt',     # Return pytorch tensors.
    )
```

```python
def get_embeddings(text):
  # Encode text and computer input_ids and attention mask
  tokens = encode(text)
  input_ids = tokens["input_ids"].to(device)
  attention_mask = tokens["attention_mask"].to(device)
  with torch.no_grad():
      output = bert_model(input_ids=input_ids, attention_mask=attention_mask)
  cls_embedding = output.last_hidden_state[:, 0, :]
  return cls_embedding
```

```python
data_scientist_cvs=cvs['preprocessed_text'][0]
front_end_developer_cvs=cvs['preprocessed_text'][1]
business_analyst_cvs=cvs['preprocessed_text'][2]
accountant_cvs=cvs['preprocessed_text'][3]
machine_learning_cvs=cvs['preprocessed_text'][4]
embed_data_scientist=get_embeddings(data_scientist_cvs)
embed_front_end_developer=get_embeddings(front_end_developer_cvs)
embed_business_analyst=get_embeddings(business_analyst_cvs)
embed_accountant=get_embeddings(accountant_cvs)
embed_machine_learning=get_embeddings(machine_learning_cvs)
embed_job_description=get_embeddings(job_description)
```

```python
def compute_cosine_similarity(a, b):
    # Move array to CPU memory, detach from computational graph which requires_grad, convert to numpy & reshape
    a = a.cpu().detach().numpy()
    b = b.cpu().detach().numpy()
    cos_sim = cosine_similarity(a, b)
    return cos_sim[0][0]
```

```python
# with preprocessing
data_scientist_similarity = compute_cosine_similarity(embed_data_scientist, embed_job_description)
front_end_developer_similarity = compute_cosine_similarity(embed_front_end_developer, embed_job_description)
business_analyst_similarity = compute_cosine_similarity(embed_business_analyst, embed_job_description)
accountant_similarity = compute_cosine_similarity(embed_accountant, embed_job_description)
machine_learning_similarity = compute_cosine_similarity(embed_machine_learning, embed_job_description)
```
